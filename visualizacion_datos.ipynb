```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\alanc\\OneDrive\\Escritorio\\Pipeline Spotify\\Top_3_Latinoamerica.csv')

```

Configuracion de estilo de Seaborn

```python
sns.set(style = 'whitegrid')
```

Grafico de Distribucion de Popularidad

```python
plt.figure(figsize=(10,6))
sns.histplot(data['Popularidad'], kde = True, bins = 30, color = 'blue')
plt.title('Distribucion de Popularidad de las Canciones')
plt.xlabel('Popularidad')
plt.ylabel('Frecuencia')
plt.show()
```

Top 3 Canciones con mayor popularidad en Latam

```python
top_3 = data.sort_values(by='Popularidad', ascending = False)

#Eliminamos los duplicados (Canciones repetidas en varios paises)

top_3 = top_3.drop_duplicates(subset='Nombre de la canciÃ³n')

top_3 = top_3.head(3)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))

bar_plot = sns.barplot(x='Popularidad', y='Nombre de la canciÃ³n', data= top_3, palette='viridis')
plt.xlabel('Popularidad')
plt.ylabel('Nombre de la cancion')
plt.title('Canciones mas Populares en Latinoamerica')
plt.show()
```

Popularidad Promedio en cada pais

```python
popularidad_por_pais = data.groupby('PaÃ­s')['Popularidad'].mean().reset_index()
popularidad_por_pais = popularidad_por_pais.sort_values(by='Popularidad', ascending= False)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
bar_plot = sns.barplot(x='Popularidad', y='PaÃ­s', data= popularidad_por_pais, palette= 'viridis', order= popularidad_por_pais['PaÃ­s'])
plt.xlabel('Popularidad')
plt.ylabel('Pais')
plt.title('Popularidad promedio de las canciones por pais en Latam')
plt.show()
```

Artistas con mas canciones en los Top

```python
artistas_conteo = data['Artista(s)'].value_counts().reset_index()
artistas_conteo.columns = ['Artista(s)', 'Numero de Canciones'] 
top_10_artistas = artistas_conteo.head(10)
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(12,6))
bar_plot = sns.barplot(x='Numero de Canciones', y='Artista(s)', data= top_10_artistas, palette= 'viridis')
plt.xlabel('Numero de Canciones')
plt.ylabel('Artista(s)')
plt.title('Top 10 Artistas con mayor numero de canciones en los ranking de Latam')
plt.show()
```
